## 4. Choose Keywords

```python
positive_keywords = [
    "excellent",
    "amazing",
    "great",
    "love"
]

negative_keywords = [
    "bad",
    "terrible",
    "awful",
    "boring"
]

print("Positive Keywords:", positive_keywords)
print("Negative Keywords:", negative_keywords)
```

## 5. Calculate Prior Probability

```python
# Total number of reviews
total_reviews = len(df)

# Count positive and negative reviews
positive_reviews = len(df[df["sentiment"] == "positive"])
negative_reviews = len(df[df["sentiment"] == "negative"])

# Calculate prior probabilities
prior_positive = positive_reviews / total_reviews
prior_negative = negative_reviews / total_reviews

print("Total Reviews:", total_reviews)
print("Positive Reviews:", positive_reviews)
print("Negative Reviews:", negative_reviews)

print("\nPrior Probability")
print("P(Positive) =", prior_positive)
print("P(Negative) =", prior_negative)
```

## 6. Calculate Likelihood

```python
# Split the dataset
positive_df = df[df["sentiment"] == "positive"]
negative_df = df[df["sentiment"] == "negative"]

print("Positive Reviews:", len(positive_df))
print("Negative Reviews:", len(negative_df))
```

```python
# Choose one keyword to test first
keyword = "amazing"

# Count how many positive reviews contain the keyword
positive_keyword_count = positive_df["review"].str.lower().str.contains(keyword).sum()

# Calculate likelihood
likelihood = positive_keyword_count / len(positive_df)

print(f"Keyword: {keyword}")
print("Occurrences in positive reviews:", positive_keyword_count)
print("Likelihood P(keyword | Positive):", likelihood)
```

## 7. Calculate Marginal Probability

```python
# Count how many reviews (positive + negative) contain the keyword
total_keyword_count = df["review"].str.lower().str.contains(keyword).sum()

# Calculate marginal probability
marginal_probability = total_keyword_count / total_reviews

print("Occurrences in all reviews:", total_keyword_count)
print("Marginal Probability P(keyword):", marginal_probability)
```

## 8. Apply Bayes' Theorem

```python
# Apply Bayes' Theorem
posterior_probability = (
    likelihood * prior_positive
) / marginal_probability

print(f"P(Positive | {keyword}) = {posterior_probability:.4f}")
```

## 9. Calculate Probabilities for All Keywords

```python
# Combine all keywords
all_keywords = positive_keywords + negative_keywords

results = []

for keyword in all_keywords:

    # Likelihood: P(keyword | Positive)
    positive_count = positive_df["review"].str.lower().str.contains(keyword).sum()
    likelihood = positive_count / len(positive_df)

    # Marginal: P(keyword)
    total_count = df["review"].str.lower().str.contains(keyword).sum()
    marginal = total_count / len(df)

    # Bayes' Theorem
    posterior = (likelihood * prior_positive) / marginal

    results.append({
        "Keyword": keyword,
        "Likelihood": round(likelihood, 4),
        "Marginal": round(marginal, 4),
        "Posterior P(Positive|Keyword)": round(posterior, 4)
    })
```

```python
results_df = pd.DataFrame(results)

results_df
```
